1. |= (Bitli **OR** va o'zlashtirish)
Bu operator ikkita sonning bitlarini solishtiradi.
- Qoida: Agar ikkita mos bitdan aqalli bittasi 1 bo'lsa, natija 1 bo'ladi. Faqatgina ikkala bit ham 0 bo'lsagina natija 0 bo'ladi.
- Ma'nosi: x |= 3 bu x = x | 3 degani.
- Misol:
    - x = 5 ➡️ 0 1 0 1
    - 3 ➡️ 0 0 1 1
- OR (|) natijasi: 0 1 1 1 (Bu o'nlikda 7 degani).
Demak, x ning qiymati 7 ga o'zgaradi.
2. ^= (Bitli **XOR** va o'zlashtirish)
Bu operatorni yuqorida ko'rib chiqdik.
- Qoida: Agar ikkita mos bit har xil bo'lsa natija 1, bir xil bo'lsa natija 0 bo'ladi.
- Ma'nosi: x ^= 3 bu x = x ^ 3 degani.
- Muhim eslatma: Matematikada ^ belgisi darajaga ko'tarishni anglatsa-da, Pythonda u faqat XOR amali hisoblanadi. Pythonda darajaga ko'tarish uchun ** ishlatiladi (masalan, x = x ** 3).
- Misol:
    - x = 5 ➡️ 0 1 0 1
    - 3 ➡️ 0 0 1 1
- XOR (^) natijasi: 0 1 1 0 (Bu o'nlikda 6 degani).
3. \>>= (Bitlarni o'ngga surish va o'zlashtirish)
Bu operator sonning ikkilik ko'rinishidagi barcha bitlarini ko'rsatilgan miqdorga o'ng tomonga suradi.
- Qoida: O'ng tomonga surilganda o'ngdagi chetki bitlar o'chib ketadi, chap tomondan esa yangi 0 lar qo'shiladi. Bu amaliyot matematik jihatdan sonni 2
n
  ga butun bo'lish bilan bir xil natija beradi (bu yerda n - surishlar soni).
- Ma'nosi: x >>= 3 bu x = x >> 3 degani (x ning bitlarini 3 xona o'ngga surish).
- Misol:
    - Keling, bu safar kattaroq son olamiz: x = 16 (ikkilikda 10000).
    - x = 16 ➡️ 1 0 0 0 0
    - 3 marta o'ngga surish: Bitlar 3 xona o'ngga siljiydi: 0 0 0 1 0 (O'ngdagi 3 ta nol o'chib ketdi).
- Natija: 2 bo'ladi. (Matematik tekshirish: 16 ni 2**3 ga, ya'ni 8 ga bo'lsak, 2 chiqadi).
4. <<= (Bitlarni chapga surish va o'zlashtirish)
Bu >>= operatorining teskarisi. Sonning bitlarini ko'rsatilgan miqdorga chap tomonga suradi.
- Qoida: Chapga surilganda o'ng tomondan 0 lar qo'shilib boradi. Bu amaliyot matematik jihatdan sonni 2
n
  ga ko'paytirish bilan bir xil natija beradi.
- Ma'nosi: x <<= 3 bu x = x << 3 degani (x ning bitlarini 3 xona chapga surish).
- Misol:
    - x = 2 (ikkilikda 00010) bo'lsin.
    - x = 2 ➡️ 0 0 0 1 0
3 marta chapga surish: Bitlar 3 xona chapga siljiydi va o'ngdan 3 ta nol qo'shiladi: 1 0 0 0 0
- Natija: 16 bo'ladi. (Matematik tekshirish: 2 ni 2
3
  ga, ya'ni 8 ga ko'paytirsak, 16 chiqadi).
